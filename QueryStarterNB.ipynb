```python
# Set up path to cafes.db
from pathlib import Path

try:
    HERE = Path(__file__).resolve().parent
except NameError:
    HERE = Path.cwd()

DB_PATH = HERE / "cafes.db"
print(DB_PATH)
```

```python
import sqlite3

conn = sqlite3.connect(DB_PATH)
conn.execute("PRAGMA foreign_keys = ON")

def run_sql(sql: str):
    try:
        cur = conn.execute(sql)
        if sql.lstrip().upper().startswith(("SELECT", "PRAGMA")):
            rows = cur.fetchall()
            return rows
        else:
            conn.commit()
            print("OK")
            return None

    except sqlite3.Error as err:
        print(f"SQLite error: {err}")
        return None
```

```python
print("Tables:")
for row in run_sql("SELECT name FROM sqlite_master WHERE type='table';"):
    print(row[0])

print("\nSchema:")
for row in run_sql("SELECT sql FROM sqlite_master WHERE type='table';"):
    print(row[0])
```

## Activity Set 2

```python
# write queries to retrieve:
# Names of the cafes
run_sql("SELECT DISTINCT cafe FROM Sells;")
# Types of drinks
run_sql("SELECT DISTINCT drink FROM Sells;")
# Names of regulars
run_sql("SELECT DISTINCT regular FROM Frequents;")
# Names of cafes that sell Iced Coffee
run_sql("SELECT cafe FROM sells WHERE drink = 'Iced Coffee'")
# Names of students who frequent Starbucks
run_sql("SELECT DISTINCT regular FROM Frequents WHERE cafe = 'Starbucks';")
# Drinks that cost more than $3
run_sql("SELECT cafe, drink, price FROM Sells WHERE price > 3")
```

## Activity Set 3

```python
#Show each drink with its price in cents
run_sql("SELECT drink, price * 100 AS priceInCents FROM Sells;")
#Show each drink with a hypothetical $1 increase
run_sql("SELECT drink, price + 1 FROM Sells;")
#Show each regular with a label
run_sql("SELECT DISTINCT regular FROM Frequents;")
#Show each drink with 6% tax added
run_sql("SELECT drink, price + (price * 0.06) FROM Sells;")
```

## Activity Set 4

```python
#Drinks at Starbucks that cost more than $3
run_sql("SELECT drink, price FROM Sells WHERE cafe = 'Starbucks' AND price > 3")
#Drinks that are either Mocha or Latte
run_sql("SELECT cafe, drink, price FROM Sells WHERE drink = 'Mocha' OR drink = 'Latte'")
#Cafes whose names start with S
run_sql("SELECT DISTINCT cafe FROM Sells WHERE cafe LIKE 'S%'")
#Cafes whose names contain the word Coffee
run_sql("SELECT DISTINCT cafe FROM Sells WHERE cafe LIKE '%Coffee%'")
#Drinks with unknown prices
run_sql("SELECT cafe, drink FROM Sells WHERE price IS NULL")
```
