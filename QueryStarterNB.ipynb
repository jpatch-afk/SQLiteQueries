```python
# Set up path to cafes.db
from pathlib import Path

try:
    HERE = Path(__file__).resolve().parent
except NameError:
    HERE = Path.cwd()

DB_PATH = HERE / "cafes.db"
print(DB_PATH)
```

```python
import sqlite3

conn = sqlite3.connect(DB_PATH)
conn.execute("PRAGMA foreign_keys = ON")

def run_sql(sql: str):
    try:
        cur = conn.execute(sql)
        if sql.lstrip().upper().startswith(("SELECT", "PRAGMA")):
            rows = cur.fetchall()
            return rows
        else:
            conn.commit()
            print("OK")
            return None

    except sqlite3.Error as err:
        print(f"SQLite error: {err}")
        return None
```

```python
print("Tables:")
for row in run_sql("SELECT name FROM sqlite_master WHERE type='table';"):
    print(row[0])

print("\nSchema:")
for row in run_sql("SELECT sql FROM sqlite_master WHERE type='table';"):
    print(row[0])
```

```python
run_sql("SELECT * FROM Sells;")
```

```python
# My favorite dataset: names sampled uniformly from SSA baby-name data.
# Better than inventing classroom placeholder people by hand.
run_sql("SELECT * FROM Frequents;")
```
